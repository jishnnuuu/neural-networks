## Reshaping, stacking, squeezing and unsqueezing tensors

* Reshaping - reshapes the input tensor into defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the orginal torch.tensor
* Stacking - combine multiple tensors on top of each other(vstack) or side by side(hstack)
* Squeeze - remove all 1 dimensions from a tensor
* Unsqueeze - add a 1 dimenstion to a target tensor
* Permute - Return a view of the input with dimensions permuted(swapped) in a certain way

```python
x = torch.arange(1., 10.)
```

```python
x, x.shape
```

```python
x_reshaped = x.reshape(3,3)
x_reshaped, x_reshaped.shape
```

```python
z = x.view(1,9)
z, z.shape
```

```python
# changing z changes x(because a view of a tensor share the same memory as the original input)

z[:,0] = 5
z,x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x,x,x,x])
x_stacked
```

```python
x_stacked = torch.stack([x,x,x,x], dim=0)
x_stacked
```

```python
x_stacked = torch.stack([x,x,x,x], dim=1)
x_stacked
```

```python
x_reshaped
```

```python
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

## Reproducibility (trying to take random out of random)

* to reduce randomness in neural networks and pytorch comes with the concept of random torch.seed

```python
random_tensorA = torch.rand(3,4)
random_tensorB = torch.rand(3,4)

print(random_tensorA)
print(random_tensorB)
print(random_tensorA == random_tensorB)
## not reproducible
```

```python
RANDOM_SEED = 42

#1
torch.manual_seed(RANDOM_SEED)
random_tensorC = torch.rand(3,4)

#2
torch.manual_seed(RANDOM_SEED)
random_tensorD = torch.rand(3,4)

print(random_tensorC)
print(random_tensorD)
print(random_tensorC == random_tensorD)
##reproducible
```
